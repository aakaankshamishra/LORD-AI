# **Data Loading and Data Cleaning**

```python
# importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loading the data
data = pd.read_csv('Day.csv')
```

```python
data.head()
```

```python
# number of rows and number of columns
data.shape
```

```python
# Statistical summary of each variables
data.describe()
```

```python
# Information regarding the variables
data.info()
```

There is one object variable and 15 numerical variable.
'cnt' is our target variable (count of total rental bikes)

```python
# checking for null values
data.isnull().sum()
```

There are no null values in our dataset

```python
# we can drop the 'instant' and 'dteday'(we have some other columns that define the dates and days) variables because they will not add any benefits for our model
data.drop(['instant', 'dteday'], axis = 1, inplace = True)
```

```python
data.shape
```

```python
data.info()
```

we need to go through the variables 'season', 'mnth', 'weekday', 'weathersit' and 'yr' for making the numbers as categorical values for better understanding.

```python
# 'season' variable
data['season'] = data['season'].map({1: 'spring', 2: 'summer', 3: 'fall', 4: 'winter'})
```

we have mapped the values 1, 2, 3, and 4 to their corresponding seasons 'spring', 'summer', 'fall', 'winter'

```python
# 'mnth' variable
data['month'] = data['mnth'].map({1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'})
```

we have mapped the values 1 to 12 to their corresponding months.

```python
# 'weathersit' variable
print(data['weathersit'].unique())
data['weathersit'] = data['weathersit'].map({1: 'Clear', 2: 'Mist + Cloudy', 3: 'Light Snow'})
```

we have mapped the values 1,2, and 3 to their corresponding weathers 'Clear', 'Mist + Cloudy', and 'Light Snow'.

```python
# 'weekday' variable
print(data['weekday'].unique())
data['weekday']=data['weekday'].map({0:'Sunday',1:'Monday',2:'Tuesday',3:'Wednesday',4:'Thursday',5:'Friday',6:'Saturday'})
```

we have mapped the values 0 to 6 to their corresponding weekdays.

```python
# 'yr' variable
data['year'] = data['yr'].map({0: 2018, 1: 2019})
```

we have mapped the values 0 and 1 to their corresponding years 2018 and 2019

```python
data
```

```python
# dropping the extra columns 'yr' and 'mnth'
data.drop(['yr', 'mnth'], axis = 1, inplace = True)
```

```python
data
```

```python
data.info()
```

```python
# found the correlaions between all variables
correlation= data.corr()
correlation
```

```python
#the feature 'atemp' is highly correlated with the feature 'temp' so we can drop it
data.drop('atemp', axis = 1, inplace = True)
data.shape
```

There is two features 'casual' and 'registered' that specifies the count of casual users and registered users. our target variable 'cnt' contains the sum of these two columns. They are highly correlated with target variable 'cnt'. Now we are not dropping these two columns, may be these columns can be useful for visualization and other tasks.

```python
# Renaming The columns for ease to read.
data.rename(columns={'hum':'humidity','cnt':'count'},inplace=True)
data.head()
```

```python
data.to_csv("cleaned_day_data.csv")
```

```python
correlations = data1.corr()['count'].abs().sort_values(ascending=False)
top_10_correlated = correlations[1:11]
print(top_10_correlated)
```

**Multicollinearity problem**(Independent vriables are correlated)

```python
correlations = data1.corr()
correlations = correlations.where(np.triu(np.ones(correlations.shape, dtype=bool), k=1))
correlations = correlations.unstack().reset_index()
correlations.columns = ['VAR1', 'VAR2', 'Correlation']
correlations.dropna(subset=["Correlation"], inplace=True)
correlations["Correlation"] = correlations["Correlation"].abs()
correlations.sort_values(by='Correlation', ascending=False, inplace=True)
correlations.head(15)
```

```python
#Dropping highly correlated  variables
data_n= data1.drop(['season_spring','workingday','registered','casual','season_winter'],axis=1)
```

### **Splitting the Data into Train and Test**

```python
# Import the required library
from sklearn.model_selection import train_test_split

```

```python
# Put all the feature variables in X
x = data_n.drop(['count'],axis=1)
x.head()
```

```python
# Put the target variable in y
y = data_n['count']
y.head()
```

### **Feature Importance check**

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(x, y)

(pd.Series(model.feature_importances_, index=x.columns)
   .nlargest(15)
   .plot(kind='barh',color='teal'))
```

From the feature importance plot we can conclude that temperature is the most important feature. Year humidity and windspeed are also the important features that contribute positively to bike rentals.

```python
# Split the dataset into 70% train and 30% test
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, test_size=0.3,random_state=100)
```

### **Model Building**

```python
# linear regression model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#importing evaluation metrics
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#prediction and evaluation
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
print("mse:", mse)
print("rmse:", rmse)
print("r2:", r2)
```

```python
#residual analysis
residuals = y_test - y_pred

# Residual plot
plt.figure(figsize=(8, 6))
plt.scatter(y_pred, residuals, alpha=0.6)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Residual Plot')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.show()
```

```python
sns.regplot(x=y_test, y=y_pred, ci=52, fit_reg=True, line_kws={"color": "red"})
plt.scatter(y_test, y_pred)
plt.suptitle('y_test vs y_pred', fontsize = 16)
plt.xlabel('y_test', fontsize = 14)
plt.ylabel('y_pred', fontsize = 14)
plt.show()
```

Inference:

As we see here the redline and the dots are in the incresing, it presents Positive linear relationship and there few outliers, model fits perfectly.

Here are some specific observations from the residual plot:

* The residuals are evenly distributed around the zero line. This means that the model is not overpredicting or underpredicting for any particular values of the independent variable.
* There is no evidence of heteroscedasticity in the residual plot. This means that the variance of the residuals is not changing systematically with the values of the independent variable.
